```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(1234)
```

# Exponential random numbers
## Define a function to generate exponential random numbers

```python
def exponential_rng(lam=1.0):
    """ Generates exponential random numbers.
    
    Keywords:
        lam (float): the rate parameter, the inverse expectation of the distribution.
    
    Returns:
        exponential random numbers with rate lam
    """
    return -(1/lam)*np.log(np.random.rand())
```

## Test the function

```python
exponential_rng()
```

## Generate draws and compare to theoretical density

```python
def exponential_pdf(x,lam=1.0):
    """Returns the exponential distribution pdf
    
    Keywords:
        x: the random variable
        lam: the exponential distribution rate parameter
        
    Returns:
        the probability density function value
    """
    return lam*np.exp(-lam*x)

lam=0.2

# Density
x=np.arange(0,25,0.01)
exp_pdf=exponential_pdf(x,lam)

# Draws
N_bins=10
draws=[exponential_rng(lam) for i in range(100)]


plt.hist(draws,N_bins,density=1,label='Draws')
plt.plot(x,exp_pdf,'r',label='Density')
plt.title("Exponential random numbers")
plt.legend()
plt.show()

# Alternative way for checking - QQ-Plot
from scipy import stats
res = stats.probplot(draws, dist=stats.expon,plot=plt)
plt.plot()
plt.show()
```

# Homogeneous Poisson process
## Define a function to generate arrivals according to Poisson process

```python
def homogeneous_poisson_process(lam, T):
    """Generates arrivals according to the homogeneous Poisson process. 
    
    Keywords:
        lam: the rate parameter
        T (int): duration of Poisson process
    
    Returns:
        arrival_times: array of event times
    """
    arrival_times=[]
    curr_arrival_time=0
    while True:
        curr_arrival_time += exponential_rng(lam)
        if curr_arrival_time > T:
            break
        else:
            arrival_times.append(curr_arrival_time)
    return arrival_times
```

## Generate homogeneous Poisson process

```python
lam=4.0
T=1
D=10000

arrivals=[homogeneous_poisson_process(lam, T) for i in range(D)]
```

## Plot arrival times

```python
flat_arrivals = [item for sublist in arrivals for item in sublist]

# Density
x=np.arange(0,25,0.01)
exp_pdf=exponential_pdf(x,lam)

N_bins=10

plt.hist(flat_arrivals,N_bins)
plt.ylabel("Arrival count")
plt.xlabel("Arrival time within Poisson process period T")
plt.title("Homogeneous Poisson process")
plt.show()
```

# Nonhomogeneous Poisson process
## Define a function to generate arrivals according to non-homogeneous Poisson process

```python
def non_homogeneous_poisson_process(lam_t, lam_max, T):
    """Generates arrivals according to the homogeneous Poisson process. 
    
    Keywords:
        lam: the rate parameter
        T (int): duration of Poisson process
    
    Returns:
        arrival_times: array of event times
    """
    arrival_times=[]
    curr_arrival_time=0
    while True:
        curr_arrival_time += exponential_rng(lam_max)
        if curr_arrival_time > T:
            break
        else:
            # Accept/Reject Method
            if np.random.rand()<=(lam_t(curr_arrival_time)/lam_max):
                arrival_times.append(curr_arrival_time)
    return arrival_times
```

## Generate nonhomogeneous poisson processes

```python
lam_t=lambda t: lam*np.sin(t)+lam
lam=4.0
lam_max=lam*2
T=10
D=1000

arrivals=[non_homogeneous_poisson_process(lam_t, lam_max, T) for i in range(D)]
```

## Plot arrival times

```python
flat_arrivals = [item for sublist in arrivals for item in sublist]

# Lamba function
t=np.arange(0,T,0.01)
lam_function=lambda t: D*lam_t(t)

N_bins=10


plt.plot(t,lam_function(t),'r',label='N*λ(t)')
plt.hist(flat_arrivals,N_bins,label='Arrivals generated')
plt.ylabel("Arrival count")
plt.xlabel("Arrival time within Poisson process period T")
plt.title("Nonhomogeneous Poisson process")
plt.legend()
plt.show()
```

## Extra questions
### Efficiency

```python
def non_homogeneous_poisson_process_extra0(lam_t, lam_max, T):
    arrival_times=[]
    curr_arrival_time=0
    counter=0
    while True:
        curr_arrival_time += exponential_rng(lam_max)
        if curr_arrival_time > T:
            break
        else:
            # Accept/Reject Method
            counter+=1
            if np.random.rand()<=(lam_t(curr_arrival_time)/lam_max):
                arrival_times.append(curr_arrival_time)
    efficiency=len(arrival_times)/counter
    return arrival_times,efficiency
```

```python
lam_t=lambda t: lam*np.sin(t)+lam
lam=4.0
lam_max=lam*2
T=10
D=1000
arrivals,efficiency=zip(*[non_homogeneous_poisson_process_extra0(lam_t, lam_max, T) for i in range(D)])
print(f'Efficiency: {np.mean(efficiency):.5f}')
```

### Piecewise lambda

```python
def non_homogeneous_poisson_process_extra1(lam_t, lam_max, T):
    """ Generates arrivals according to non-homogeneous Poisson process with time-dependent rate lam_t. 
    
    Keywords:
        lam_t (function handle): function returning the Poisson process rate at time t
        lam_max (real): maximum value of lam_t(t) to be used for generating arrivals
        T (int): duration of Poisson process
    
    Returns:
        times: array of event times
        efficiency (real): (0-1) efficiency of non-homogeneous Poisson process 
    """
    arrival_times=[]
    curr_arrival_time=0
    counter=0
    
    #Piecewise lam_max
    lam_max_tmp=lam_max
    lam_max_tmp_next=lam_max/2
    interval=1
    
    while True:
        inter_arrival_time= exponential_rng(lam_max_tmp)
        while True:
            if curr_arrival_time + inter_arrival_time > interval*np.pi:
                if curr_arrival_time + inter_arrival_time > T:
                    return arrival_times,len(arrival_times)/counter
                else:
                    inter_arrival_time -=interval*np.pi-curr_arrival_time
                    inter_arrival_time *=lam_max_tmp / lam_max_tmp_next
                    curr_arrival_time=interval*np.pi

                    interval+=1
                    lam_max_tmp_old=lam_max_tmp
                    lam_max_tmp=lam_max_tmp_next
                    lam_max_tmp_next=lam_max_tmp_old
            
            else:
                if curr_arrival_time+inter_arrival_time>T:
                    return arrival_times,len(arrival_times)/counter
                else:
                    break
        curr_arrival_time+=inter_arrival_time
        counter+=1
        if np.random.rand()<=(lam_t(curr_arrival_time)/lam_max_tmp):
            arrival_times.append(curr_arrival_time)
```

```python
arrivals,efficiency=zip(*[non_homogeneous_poisson_process_extra1(lam_t, lam_max, T) for i in range(D)])
print(f'Efficiency: {np.mean(efficiency):.5f}')
```
